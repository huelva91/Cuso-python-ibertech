-Monitor  
+id

+precio

-Cliente  
+nombre
+tipo (VIP o no)
+carrito_compra: listado[productos]
+pagar()

-Cajera  
+numero_empleado  
+nombre  
+atender_cliente()  

-Tienda  
3 clientes
2 cajeras  

Cada cliente va llenar la cesta con uno o varios productos  

cajeras van a ir atendiendo a los clientes  


Si cliente es VIP, 15% en la compra

```python
class Television:
    def __init__(self, id, marca, pulgadas, precio):
        self.id = id
        self.marca = marca
        self.pulgadas = pulgadas
        self.precio = precio
    def __str__(self):
        return "Marca:" + self.marca + ", precio" + self.precio


class Ordenador:
    def __init__(self, id, marca, capacidad_disco_duro, ram, precio):
        self.id = id
        self.marca = marca
        self.capacidad_disco_duro = capacidad_disco_duro
        self.ram = ram
        self.precio = precio


class Monitor:
    def __init__(self, id, marca, pulgadas, precio):
        self.id = id
        self.marca = marca
        self.pulgadas = pulgadas
        self.precio = precio


class Cliente:
    def __init__(self, nombre, tipo, vip, carrito_compra=[]):
        self.nombre = nombre
        self.tipo = tipo
        self.vip = vip
        self.carrito_compra = carrito_compra

    def pagar(self, importe):
        print("El cliente {} esta pagando un total de {}".format(self.nombre, importe))

class Cajera:
    def __init__(self, numero_empleado, nombre):
        self.numero_empleado = numero_empleado
        self.nombre = nombre

    def atender_cliente(self, cliente):
        importe = 0
        for producto in cliente.carrito_compra:
            importe += producto.precio

        if cliente.vip:
            importe-=(importe*0.15)

        print("El cliente {} tiene que pagar {} euros".format(cliente.nombre, importe))

        return importe

class Tienda:
    def __init__(self, nombre, cajeras, clientes = []):
        self.nombre = nombre
        self.cajeras = cajeras
        self.clientes = clientes
    def add_cliente(self, cliente):
        self.clientes.append(cliente)

    def add_cajeras(self, cajera):
        self.cajeras.append(cajera)

    def cobras_clientes(self):
        hay_cliente = self.hay_clientes()
        while hay_cliente:
            for cajera in self.cajeras:
                if(hay_cliente):
                    cliente = self.clientes.pop(0)
                    print("La cajera {} esta atendiendo al cliente {}".format(cajera.nombre, cliente.nombre))
                    cajera.atender_cliente(cliente)
                    hay_cliente = self.hay_clientes()

    def hay_clientes(self):
        return len(self.clientes)>0




television1 = Television("1T", "LG", 55, 1000)
ordenador1 = Ordenador("1o", "HP", 1000, 8, 850)
monitor1 = Monitor("1M", "Samsung", 32, 400)


carrito_compra1 = [television1, ordenador1, monitor1]
carrito_compra2 = [television1, ordenador1, monitor1]
carrito_compra3 = [television1, ordenador1, monitor1]


cliente1 = Cliente("Ana", True, carrito_compra1 )
cliente2 = Cliente("Paco", False, carrito_compra2 )
cliente3 = Cliente("Lucia", True, carrito_compra3 )

cajera1 = Cajera(1, "Maria" )
cajera2 = Cajera(2, "Sergio")


lista_clientes = [cliente1, cliente2, cliente3]
lista_cajeras = [cajera1, cajera2]

tienda1 = Tienda("Supermecado 1", [cajera1, cajera2])
#Anadimos clientes a la cola de la caja

tienda1.add_cliente(cliente1)
tienda1.cobras_clientes()







```
